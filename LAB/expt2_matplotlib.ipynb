## 1. Box plot

Box plots are the summary plots of the data distribution giving information like spread, skewness, whiskers, meadian and mean.

### Box Plot

```python
import matplotlib.pyplot as plt
help(plt.boxplot)
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Creating dataset
# The np.random.seed(1) function is a useful tool for debugging 
# and for ensuring that your code produces the same results each time it is run.
#np.random.seed(1.0) 

data_1 = np.random.normal(loc= 100, scale = 10, size =200) # mean, std, size
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]

# boxplot
plt.figure(figsize = (8,6))
plt.boxplot(data, labels=['data_1', 'data_2', 'data_3','data_4'], 
            showmeans=True, 
            patch_artist=True, 
            vert=True)
xlabel = plt.xlabel('random numbers')
ylabel = plt.ylabel('group')
title = plt.title('Boxplot')


# Inserting text - showing quartiles
for i in range(0,4):
  Q1 = pd.DataFrame(data[i]).quantile(0.25)
  Q3 = pd.DataFrame(data[i]).quantile(0.75)
  plt.text(i+1, Q3+2, s = 'Q3(75%)', fontsize = 8)
  plt.text(i+1, Q1-7, s = 'Q1(25%)', fontsize = 8)
  median = pd.DataFrame(data[i]).median()
  plt.text(i+1, median, s = "median", fontsize = 8)
  
# show plot 
plt.show()
```

```python
# replication

np.random.seed = 42
data1 = np.random.normal(loc = 100, scale = 10, size =200)
data2 = np.random.normal(90,8,200)
data3 = np.random.normal(80,7,200)
data4 = np.random.normal(70,6,200)
data = [data1,data2,data3,data4]
plt.boxplot(data, labels = ['d1','d2','d3','d4'],
            vert = True, showmeans = True, patch_artist=True, notch=True)
plt.xlabel("datas")
plt.ylabel("spread")
plt.show()
```

#### Finding Outliers

```python
import pandas as pd
data1 = pd.DataFrame(data_1)

# Inter Quartile Range
Q1 = data1.quantile(0.25)
Q3 = data1.quantile(0.75)
IQR = Q3 - Q1

# Outliers
outlier = data1[(data1 < (Q1 - 1.5 * IQR)) | (data1 > (Q3 + 1.5 * IQR))].dropna()
print(outlier)
```

## 3. Histogram Plot

A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval.

```python
from matplotlib import pyplot as plt
import numpy as np

# Creating dataset
a = np.array([22, 87, 5, 43, 56,
      73, 55, 54, 11,
      20, 51, 5, 79, 31,
      27])

# Creating histogram
fig, ax = plt.subplots(figsize =(8, 6))
ax.hist(a, bins = [0, 25, 50, 75, 100]) #The bins are evenly spaced from 0 to 100, with a bin width of 25. 
# The height of each bar in the histogram represents the number of data points that fall within that bin.
plt.xlabel("bins")
plt.ylabel("frequency of values")
plt.title("Histogram")

# Show plot
plt.show()
```

## 4. Bar Chart

```python
# creating the dataset
# dictionary
data = {'C':20, 'C++':15, 'Java':30, 'Python':35}

#get the keys and values as separate lists
courses = list(data.keys())
values = list(data.values())

# creating the bar plot
plt.figure(figsize = (8, 6))
plt.bar(courses, values, color =['orange','blue','green','purple'], width = 0.4)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
