# Binary Classification

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import make_blobs
# Generate isotropic Gaussian blobs for clustering.
X, y, centers = make_blobs(n_samples = 1000, n_features = 2, centers = 2, return_centers = True, random_state = 42)
print("X shape: ",X.shape)
print("y shape: ",y.shape)
print("centers: ", centers)
```

```python
count = Counter(y)
print("from counter: ",count)
val_cnt = pd.value_counts(y)
print("Value count:\n" ,val_cnt)
```

```python
from seaborn.widgets import color_palette
import seaborn as sns
sns.scatterplot(x = X[:,0], y = X[:,1], hue = y, palette = 'viridis')
plt.scatter(centers[:,0],centers[:,1])
plt.title("Scatter plot for binary classification")
plt.xlabel("feature1")
plt.ylabel("feature2")
```

```python
def model_predict(X, y):
    # check classes
    if len(pd.Series(y).unique()) > 2:
        print("Multiclass Classification: ")
    else:
        print("Binary Classification: ")


    # split dataset
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)

    # model train and test
    from sklearn.linear_model import LogisticRegression
    lr = LogisticRegression()
    lr.fit(X_train, y_train)
    y_hat = lr.predict(X_test)

    # accuracy
    from sklearn.metrics import accuracy_score
    acc = accuracy_score(y_test, y_hat).round(3) * 100
    print("Accuracy: ", acc, "%")

    return acc

model_predict(X, y)
```

# Multiclass Classification

```python
X, y, centers  = make_blobs(n_samples = 1000, n_features = 10, centers = 6, random_state= 42, return_centers = True)
print("X shape: ",X.shape)
print("y shape: ",y.shape)
print("centers: ", centers)
```

```python
from seaborn.widgets import color_palette
import seaborn as sns
sns.scatterplot(x = X[:,0], y = X[:,1], hue = y, palette = 'viridis')
plt.scatter(centers[:,0],centers[:,1])
plt.title("Scatter plot for multiclass classification")
plt.xlabel("feature1")
plt.ylabel("feature2")
```

```python
model_predict(X, y)
```
