### Basics of Classes

**A class** is a user-defined blueprint or prototype from which objects are created. Classes provide a means of bundling data and functionality together. Each class instance can have attributes attached to it for maintaining its state. Class instances can also have methods (defined by their class) for modifying their state.

**Object of Python Class**
An Object is an instance of a Class. A class is like a blueprint while an instance is a copy of the class with actual values. It’s not an idea anymore, it’s an actual dog, like a dog of breed pug who’s seven years old. You can have many dogs to create many different instances, but without the class as a guide, you would be lost, not knowing what information is required.

An object consists of:

- State: It is represented by the attributes of an object. It also reflects the properties of an object.
- Behavior: It is represented by the methods of an object. It also reflects the response of an object to other objects.
- Identity: It gives a unique name to an object and enables one object to interact with other objects.

### Class and Instance Variables
- Instance variables are for data, unique to each instance. 
- Instance variables are variables whose value is assigned inside a constructor or method with self.
- Class variables are for attributes and methods shared by all instances of the class. 
- Class variables are variables whose value is assigned in the class.

eg5: Get Dataset class:

```python
class GetRandomDataSet:

    # class variable: shared by all instances
    X = pd.DataFrame([[1,2,3,4,5],
                  [23, np.NaN,2,4,5],
                  [3,23,np.NaN, np.NaN],
                  [2,4,6,7,8],
                  [546,352,55,66,7]],
                  columns=['f1','f2','f3','f4','f5'])
    
    # class variable
    y = pd.Series([1,2,3,4,5])

    # constructor
    def __init__(self, no_of_samples = 100 , no_of_features = 15):
        # instance variables: unique to each instance
        self.s = no_of_samples 
        self.f = no_of_features

    def get_default_X(self):
        # return the class variable
        return GetRandomDataSet.X, GetRandomDataSet.y

    def get_X_y(self, include_nan=True):
        from sklearn.datasets import make_classification
        mk_class = make_classification
        X, y = mk_class(n_samples = self.s, 
                        n_features = self.f, 
                        n_redundant=2, 
                        n_classes = 3, 
                        n_informative=3, 
                        random_state=1)
        
        # make some of the values NaN when include_nan = True
        if include_nan == True:
            rand_list = []
            for i in range(6):
                r = np.random.randint(low = 0, high = self.s)
                if r not in rand_list:
                    rand_list.append(r)
            for i in rand_list:
                for j in range(self.f):
                    X[i,j] = np.NaN 

        # convert the numpy array to pandas dataframe
        X = pd.DataFrame(X)
        X.columns = ['f'+str(i) for i in range(1,self.f+1)] # add column names
        y = pd.Series(y)
        y.name = 'target'
        return X, y
    
    
# class should be ended with () to call the constructor
# result - 1: with class parameters
get_data1 = GetRandomDataSet(no_of_samples=50,no_of_features=10) 
X1, y1 = get_data1.get_X_y(include_nan=True)
print("X1: \n", X1.head(3), "\n")

# result - 2: without class parameters - default values will be used
get_data2 = GetRandomDataSet()
X2, y2 = get_data2.get_X_y(include_nan=True)
print("X2: \n", X2.head(3), "\n")

# result - 3: get_default_X() method
X3, y3 = get_data2.get_default_X()
print("X3: \n", X3.head(), "\n")
print("y3: \n", y3.head(), "\n")
```

## Inheritance
- Inheritance is the capability of one class to derive or **inherit the properties** from another class. 
- It is a mechanism that allows you to create a **hierarchy** of classes that share a set of properties and methods by deriving a class from another class. 
- The class from which properties and methods are inherited is called the **Parent or Super class**.

**Benefits of inheritance are:**

- It represents real-world relationships well.
- It provides the **reusability** of a code. We don’t have to write the same code again and again. Also, it allows us to add more features to a class without modifying it.
- It is transitive in nature, which means that **if class B inherits from another class A, then all the subclasses of B would automatically inherit from class A.**
- Inheritance offers a simple, understandable model structure. 
- Less development and maintenance expenses result from an inheritance.

```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from scipy.stats import zscore, iqr

class DatasetInfo:
    def __init__(self, data):
        self.data = data
    
    def get_info(self):
        info = {
            "NaN Count": self.data.isnull().sum().sum(),
            "Duplicate Count": self.data.duplicated().sum(),
            "Scaled": isinstance(self.data.iloc[0, 0], (int, float)),
            "Outlier Count (Z-score)": self._count_outliers_zscore(),
            "Outlier Count (IQR)": self._count_outliers_iqr()
        }
        return info
    
    def _count_outliers_zscore(self, threshold=3):
        z_scores = np.abs(zscore(self.data))
        return np.sum(z_scores > threshold)
    
    def _count_outliers_iqr(self, whisker=1.5):
        Q1 = self.data.quantile(0.25)
        Q3 = self.data.quantile(0.75)
        IQR = iqr(self.data)
        lower_bound = Q1 - whisker * IQR
        upper_bound = Q3 + whisker * IQR
        return np.sum((self.data < lower_bound) | (self.data > upper_bound))

class DropNanTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        return X.dropna()

class DropOutliersTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, method="zscore", threshold=3, whisker=1.5):
        self.method = method
        self.threshold = threshold
        self.whisker = whisker
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        if self.method == "zscore":
            z_scores = np.abs(zscore(X))
            return X[(z_scores <= self.threshold).all(axis=1)]
        elif self.method == "iqr":
            Q1 = X.quantile(0.25)
            Q3 = X.quantile(0.75)
            IQR = iqr(X)
            lower_bound = Q1 - self.whisker * IQR
            upper_bound = Q3 + self.whisker * IQR
            return X[((X >= lower_bound) & (X <= upper_bound)).all(axis=1)]
        else:
            raise ValueError("Invalid outlier detection method specified.")

class ImputeNanTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, strategy="mean"):
        self.strategy = strategy
        self.imputer = SimpleImputer(strategy=self.strategy)
    
    def fit(self, X, y=None):
        self.imputer.fit(X)
        return self
    
    def transform(self, X, y=None):
        return pd.DataFrame(self.imputer.transform(X), columns=X.columns)

class ImputeOutliersTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, method="mean", n_neighbors=5):
        self.method = method
        self.n_neighbors = n_neighbors
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        if self.method == "mean":
            Q1 = X.quantile(0.25)
            Q3 = X.quantile(0.75)
            IQR = iqr(X)
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            X = X.mask((X < lower_bound) | (X > upper_bound)).fillna(X.mean())
        elif self.method == "knn":
            knn_imputer = KNeighborsRegressor(n_neighbors=self.n_neighbors)
            nan_mask = X.isnull()
            X = pd.DataFrame(
                knn_imputer.fit_transform(X),
                columns=X.columns,
                index=X.index
            )
            X[nan_mask] = np.nan
        elif self.method == "iterative":
            imputer = IterativeImputer()
            X = pd.DataFrame(
                imputer.fit_transform(X),
                columns=X.columns,
                index=X.index
            )
        else:
            raise ValueError("Invalid outlier imputation method specified.")
        
        return X

class ScalingTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, method="minmax"):
        self.method = method
        if self.method == "minmax":
            self.scaler = MinMaxScaler()
        elif self.method == "standard":
            self.scaler = StandardScaler()
    
    def fit(self, X, y=None):
        self.scaler.fit(X)
        return self
    
    def transform(self, X, y=None):
        return pd.DataFrame(self.scaler.transform(X), columns=X.columns)

class DataProcessor:
    def __init__(self, data):
        self.data = data
    
    def preprocess_data(self):
        preprocessing_pipeline = Pipeline([
            ("drop_nan", DropNanTransformer()),
            ("drop_outliers", DropOutliersTransformer()),
            ("impute_nan", ImputeNanTransformer()),
            ("impute_outliers", ImputeOutliersTransformer()),
            ("scaling", ScalingTransformer())
        ])
        
        self.data = preprocessing_pipeline.fit_transform(self.data)

# Example usage
data = {'A': [1, 2, 3, np.nan, 5],
        'B': [6, 7, 8, 9, 10]}
df = pd.DataFrame(X2)

info = DatasetInfo(df)
print(info.get_info())

processor = DataProcessor(df)
processor.preprocess_data()
cleaned_data = processor.data
print(cleaned_data.head())
```
